```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv('Books.csv')
ratings = pd.read_csv('Ratings.csv')
Users = pd.read_csv('Users.csv')
```

```python
books.head()
```

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
num_ratings = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings.columns = ['Book-Title', 'Num_Ratings']
num_ratings.head()
```

```python
avg_ratings = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings.columns = ['Book-Title', 'avg_Ratings']
avg_ratings.head()
```

```python
popularity = pd.merge(num_ratings, avg_ratings, on='Book-Title')
```

```python
popularity.head()
```

```python
popularity =popularity[popularity['Num_Ratings'] > 100].sort_values('avg_Ratings', ascending=False).head(50)
```

```python
popularity = popularity.merge(books, on='Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Publisher','Image-URL-M']]
```

```python
popularity.drop_duplicates('Book-Title', inplace=True)
```

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_users=x[x].index
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
final_rating= filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_rating.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt.fillna(0, inplace=True)
```

```python
pt.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim = cosine_similarity(pt)
```

```python
def recommend(title, cosine_sim=cosine_sim):
    idx = np.where(pt.index == title)[0][0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    for i in sim_scores:
        print(pt.index[i[0]])
        poster = books[books['Book-Title'] == pt.index[i[0]]]['Image-URL-M'].values[0]
        print(poster)
```

```python
recommend('Animal Farm')
```
