```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credit = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credit, on='title')
```

```python
movies= movies[['title' ,'genres','id', 'overview', 'keywords' ,'cast', 'crew']]
movies.head()
```

```python
movies.isnull().sum()
movies.dropna(inplace=True)
movies.duplicated().sum()
movies.drop_duplicates('title',inplace=True)
```

```python
movies.genres
```

```python
import ast
```

```python
def covert(obj):
    list=[]
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list
```

```python
movies['genres'] = movies['genres'].apply(covert)
movies['keywords']= movies['keywords'].apply(covert)
movies['cast']= movies['cast'].apply(covert)
movies.head()
```

```python
def fetch_director(obj):
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
movies['director'] = movies['crew'].apply(fetch_director)
movies = movies.drop('crew', axis=1)
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['director'] = movies['director'].astype('str').apply(lambda x:[x.replace(' ', '')])
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['director']
movies.head()
```

```python
movies = movies.drop('genres', axis=1)
movies = movies.drop('overview', axis=1)
movies = movies.drop('keywords', axis=1)
movies = movies.drop('cast', axis=1)
movies = movies.drop('director', axis=1)
movies.head()
```

```python
movies['tags']= movies['tags'].apply(lambda x: " ".join(x))
movies.head()
```

```python
movies['tags'][0]
```

```python
movies['tags'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    return ' '.join([ps.stem(word) for word in text.split()])
```

```python
movies['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000 , stop_words='english')
```

```python
vector = cv.fit_transform(movies['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommend(title):
    idx = movies[movies['title'] == title].index[0]
    sim_scores = list(enumerate(similarity[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]

    recd_movies = []
    recd_movies_poster = []
    for i in sim_scores:
        movie_id = recd_movies.append(movies.iloc[i[0]].id)
        try:
            recd_movies.append(movies.iloc[i[0]].title)
        except:
            pass
    return recd_movies, recd_movies_poster
```

```python
recommend('Batman Begins')
```
