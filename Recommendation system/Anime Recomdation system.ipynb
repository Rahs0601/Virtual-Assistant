```python
anime = pd.read_csv('anime.csv')
rating = pd.read_csv('rating.csv')
```

```python
anime = anime[['anime_id', 'name', 'genre', 'rating']]
```

```python
anime.head()
```

```python
anime['genre'] = anime['genre'].astype(str).apply(lambda x: x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    return ' '.join([ps.stem(word) for word in text.split()])
```

```python
anime['genre'] = anime['genre'].apply(stem)
anime['genre'].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000 , stop_words='english')
```

```python
vector = cv.fit_transform(anime['genre']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
def get_recommendations(title):
    index = anime[anime['name'] == title].index[0]
    sim_scores = list(enumerate(similarity[index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    anime_indices = [i[0] for i in sim_scores]
    anime_name = anime.iloc[anime_indices]['name']
    return anime_name
```

```python
get_recommendations('Gintama&#039;')
```

```python
import pickle 
pickle.dump(anime.to_dict(), open('Anime.pkl','wb'))
pickle.dump(similarity, open('Anime_similarity.pkl','wb'))
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
def recommend_anime(title):
    idx = anime[anime['name'] == title].index[0]
    sim_scores = list(enumerate(similarity[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]

    recd_anime = []
    recd_anime_poster = []
    for i in sim_scores:
        anime_id = anime.iloc[i[0]]['anime_id']
        recd_anime.append(anime.iloc[i[0]].name)
        recd_anime_poster.append(fetch_poster_anime(anime_id))
    return recd_anime, recd_anime_poster


def fetch_poster_anime(anime_id):
    htmldata = urlopen(f'https://myanimelist.net/anime/{anime_id}')
    soup = BeautifulSoup(htmldata, 'html.parser')
    images = soup.find_all('img')
    link = images[1]['data-src']
    return link
```

```python
recommend_anime('Gintama&#039;')
```
